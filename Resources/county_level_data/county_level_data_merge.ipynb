```python
import pandas as pd
```

```python
coc_fips_df = pd.read_csv('county_coc_source.csv')
coc_fips_df.head(10)
```

```python
population_county_df = pd.read_csv('co-est2019-annres_clean.csv')
population_county_df.head(10)
```

```python
county_coc_fips_df = pd.merge(coc_fips_df, population_county_df, left_on=['FIPS'], right_on=['FIPS'], how='left')
county_coc_fips_df.head(10)
```

```python
unemployment_county_df = pd.read_csv('unemployment_data_clean.csv')
unemployment_county_df.head(10)
```

```python
county_main_df = pd.merge(county_coc_fips_df, unemployment_county_df, left_on=['FIPS'], right_on=['FIPS_Code'], how='left')
county_main_df.head(10)
```

```python
county_main_df.to_csv("county_data.csv", index=False)
```
